```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

D = 0.4
u = 0.8
t_init = 0
t_stop = 4.
x_left = 0
x_right = 10
Nt = 100
Nx = 240

fig, ax = plt.subplots(3, 2, figsize=(10, 10))
ax = ax.ravel()
fig.set_tight_layout(True)


def T_fn(x, t):
    return np.exp(-(x - 5.5 - u * t) ** 2 / D) \
           + np.heaviside(x - 1 - u * t, 0.5) * np.heaviside(-(x - 2 - u * t), 0.5) \
           + np.minimum(np.maximum(x - 3 - u * t, 0), np.maximum(-(x - 4 - u * t), 0)) * 2


def T_init_num(x):
    return T_fn(x, t=t_init)


xx = np.linspace(x_left, x_right, Nx + 1, endpoint=True)
tt = np.linspace(t_init, t_stop, Nt + 1, endpoint=True)
dx = xx[-1] - xx[-2]
dt = tt[-1] - tt[-2]
r = u * dt / dx
t_table_record = []
schemes = ['unstable', 'upwind', 'leapfrog', 'lax_friedrichs',
           'lax_wendroff', 'beam_warming']
for idx, scheme in enumerate(schemes):

    t_table = np.zeros((len(tt), len(xx)))
    t_table[0, :] = T_init_num(xx)

    if scheme == 'upwind':
        for i in range(len(t_table) - 1):
            t_table[i + 1, 1:] = (1 - r) * t_table[i, 1:] \
                                 + r * t_table[i, :-1]
            t_table[i + 1, 0] = T_fn(x=x_left, t=dt * i)

    elif scheme == 'lax_friedrichs':
        for i in range(len(t_table) - 1):
            t_table[i + 1, 1:-1] = 0.5 * (1 - r) * t_table[i, 2:] \
                                   + 0.5 * (1 + r) * t_table[i, :-2]
            t_table[i + 1, 0] = T_fn(x=x_left, t=dt * i)
            t_table[i + 1, -1] = T_fn(x=x_right, t=dt * i)

    elif scheme == 'lax_wendroff':
        for i in range(len(t_table) - 1):
            t_table[i + 1, 1:-1] = 0.5 * (r - 1) * r * t_table[i, 2:] \
                                   + (1 - r * r) * t_table[i, 1:-1] \
                                   + 0.5 * (1 + r) * r * t_table[i, :-2]
            t_table[i + 1, 0] = T_fn(x=x_left, t=dt * i)
            t_table[i + 1, -1] = T_fn(x=x_right, t=dt * i)

    elif scheme == 'beam_warming':
        for i in range(len(t_table) - 1):
            t_table[i + 1, 2:] = (1 - 1.5 * r + r * r * 0.5) * t_table[i, 2:] \
                                 + (2 * r - r * r) * t_table[i, 1:-1] \
                                 + (-0.5 * r + 0.5 * r * r) * t_table[i, :-2]
            t_table[i + 1, 0] = T_fn(x=x_left, t=dt * i)
            t_table[i + 1, 1] = T_fn(x=x_left + dx, t=dt * i)

    elif scheme == 'leapfrog':
        t_table[1, :] = T_fn(x=xx, t=dt)
        for i in range(1, len(t_table) - 1):
            t_table[i + 1, 1:-1] = -r * t_table[i, 2:] \
                                   + t_table[i - 1, 1:-1] \
                                   + r * t_table[i, :-2]
            t_table[i + 1, 0] = T_fn(x=x_left, t=dt * i)
            t_table[i + 1, -1] = T_fn(x=x_right, t=dt * i)

    else:
        for i in range(len(t_table) - 1):
            t_table[i + 1, 1:-1] = -0.5 * r * t_table[i, 2:] \
                                   + t_table[i, 1:-1] \
                                   + 0.5 * r * t_table[i, :-2]
            t_table[i + 1, 0] = T_fn(x=x_left, t=dt * i)
            t_table[i + 1, -1] = T_fn(x=x_right, t=dt * i)

    t_table_record.append(t_table)
idx_frame = Nt


def update(idx_frame):
    for idx, t_table in enumerate(t_table_record):
        ax[idx].clear()
        ax[idx].plot(xx, t_table[idx_frame, :])
        ax[idx].plot(xx, T_fn(t=idx_frame * dt + t_init, x=xx),'--')
        ax[idx].set_xlim([-0.5, 10.5])
        ax[idx].set_ylim([-0.4, 1.3])
        ax[idx].set_xticks([])
        ax[idx].set_yticks([])
        ax[idx].set_title(schemes[idx])

update(Nt)
plt.show()
plt.close()
```
