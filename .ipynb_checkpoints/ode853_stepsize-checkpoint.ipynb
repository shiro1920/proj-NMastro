```python
import numpy as np
from scipy.integrate import ode
from scipy.integrate import solve_ivp
import matplotlib 
import matplotlib.pyplot as plt
```

```python
def func_h(z,f,S0,R200,Rs,r,a_g,b_g):
    H0,H1 = f
    dH0dz = H1
    
    R200 = R200*1000
    Rs   = Rs*1000
    rhoc = 1.4789927*10**(-7)
    c    = R200/Rs
    g    = 1/(np.log(1+c)-c/(1+c))
    derta= 200*c**3*g/3
    rho0 = rhoc*derta
    R1   = np.sqrt(r**2+z**2)
    p_DM = rho0/((R1/Rs)*(1+R1/Rs)**2)  
    
    dH1dz = H1**2/H0-a_g*H0*(H0+S0+p_DM)+b_g*H0    
    return [dH0dz,dH1dz]
```

```python
z0 = 0
y0 = [0.007,0]
z_bound = 6000
args = [0.004,65,44,670,0.002669,6.35e-6]
```

## 不设置初始步长

```python
solver = solve_ivp(func_h,t_span=[z0,z_bound],y0=y0,args=args,method='DOP853')
```

```python
plt.plot(solver.t,solver.y[0],'.',color='r',label='H0(z)')
plt.plot(solver.t,solver.y[1],'.',color='b',alpha=0.5,label='H1(z)')
plt.legend(loc='best')
```

```python
np.diff(solver.t)
```

## 设置初始步长，不设置上下限

### 初始步长为10

```python
solver = solve_ivp(func_h,t_span=[z0,z_bound],y0=y0,args=args,method='DOP853',\
                   first_step=10)
```

```python
plt.plot(solver.t,solver.y[0],'.',color='r',label='H0(z)')
plt.plot(solver.t,solver.y[1],'.',color='b',alpha=0.5,label='H1(z)')
plt.legend(loc='best')
```

```python
np.diff(solver.t)
```
